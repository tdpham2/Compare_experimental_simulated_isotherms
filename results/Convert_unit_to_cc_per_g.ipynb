```python
from ase.io import read
import json
import os
from collections import defaultdict

cifpath = '/home/tdpham/software/Compare_experimental_simulated_isotherms/setup_simulations/CIFs'
cifassign = '/home/tdpham/software/Compare_experimental_simulated_isotherms/setup_simulations/cif_assignment.txt'
path_to_isotherms = '/home/tdpham/software/Compare_experimental_simulated_isotherms/csd/matched_CIFs_and_DOI' 
cif_assignment = []
iso_assignment = []

adsorbate_mol_weight = {'Nitrogen': 28, 'Carbon Dioxide': 44, 'Methane': 16, 'Hydrogen': 2}

with open(cifassign, 'r') as f:
    for l in f:
        line = l.strip().split(',')
        iso_assignment.append(line[0])
        cif_assignment.append(line[1])

output_dict = {}

for iso_ind, iso in enumerate(iso_assignment):
    doi = iso.lower().split('isotherm')[0][:-1].lower()
    isoname = iso[:-5]
    path_to_iso = os.path.join(path_to_isotherms, doi, iso)
    with open(path_to_iso, 'r') as f:
        iso_data = json.load(f)
    unit = iso_data['adsorptionUnits']
    adsorbate_name = iso_data['adsorbates'][0]['name']
    isotherm_data = iso_data['isotherm_data']
    temperature = iso_data['temperature']
    pressures = []
    uptakes = []
    for ind, dat in enumerate(isotherm_data):
        pressures.append(dat['pressure'])
        uptakes.append(dat['species_data'][0]['adsorption'])
        converted_pressures = [round(i*100000, 1) for i in pressures]

    if unit == 'cm3(STP)/g' or unit == 'ml/g':
        print('cc/g')
        converted_uptakes = uptakes

    elif unit == 'mmol/g' or unit == 'mol/kg':
        converted_uptakes = [round(i*22.4, 3) for i in uptakes]

    elif unit == 'wt%':
        converted_uptakes = [i/(100-i)/adsorbate_mol_weight[adsorbate_name]*22.4*1000 for i in uptakes]
    
    elif unit == 'mg/g':
        converted_uptakes = [i*22.4/adsorbate_mol_weight[adsorbate_name] for i in uptakes]
    else:
        print(isoname)
        
    if doi in output_dict:
        output_dict[doi][isoname] = {}
        
    else:
        output_dict[doi] = {}
        output_dict[doi][isoname] = {}
    output_dict[doi][isoname]['isotherm_data'] = []
    output_dict[doi][isoname]['temperature'] = temperature
    output_dict[doi][isoname]['adsorbate'] = adsorbate_name
    output_dict[doi][isoname]['adsorbent'] = cif_assignment[iso_ind]
    for i, j in zip(converted_pressures, converted_uptakes):
        data_to_add = {'pressure': i, 'uptake': j}
        output_dict[doi][isoname]['isotherm_data'].append(data_to_add)
        
with open('converted_unit.json', 'w') as f:
    json.dump(output_dict, f, indent=4)
```
